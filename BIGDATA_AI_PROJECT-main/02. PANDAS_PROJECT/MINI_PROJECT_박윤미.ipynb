### [1] 데이터 준비
- DF 로딩 / 기본 정보 확인

```python
file='서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv'
```

```python
# DF 로딩
# 데이터 파일 중 2022년 봄에 관한 데이터 추출
subway=pd.read_csv(file,encoding='cp949')
subway_spring=(subway['사용월']>=202203) & (subway['사용월']<=202205)
subway_2022_spring=subway.loc[subway_spring]
```

```python
# 기본 정보 확인
subway_2022_spring.info()
```

```python
# info와 실제 데이터의 타입 매칭 체크
subway_2022_spring.head()
```

```python
# 수치 데이터의 컬럼별 분포 확인
subway_2022_spring.describe()
```

### [2] 데이터 전처리
- 결측치 체크 / 중복 체크

```python
subway_2022_spring.isnull().sum()
```

#### 중복 체크

```python
subway_2022_spring.duplicated().sum()
```

#### 승하차별 합계 구하기

```python
# 승차 합계
subway_in=subway_2022_spring.T.iloc[3:-1:2].sum()
# 하차 합계
subway_out=subway_2022_spring.T.iloc[4:-1:2].sum()
```

```python
subway_2022_spring=subway_2022_spring.T.copy()[0:3]
```

#### 승차인원과 하차인원 합계를 데이터프레임에 합치기

```python
subway_2022_spring.loc['승차인원 합계']=subway_in
subway_2022_spring.loc['하차인원 합계']=subway_out
```

#### 각 월별 데이터프레임 만들기

```python
subway_2022_spring=subway_2022_spring.T
subway_2022_spring=subway_2022_spring.set_index('사용월')

# 3월
subway_03=subway_2022_spring.loc[202203]
# 4월
subway_04=subway_2022_spring.loc[202204]
# 5월
subway_05=subway_2022_spring.loc[202205]
```

```python
subway_03
```

#### 총 합계를 기준으로 교통 혼잡도 구간 분할

```python
traffic_03=subway_03.copy()
traffic_04=subway_04.copy()
traffic_05=subway_05.copy()
```

```python
# 월별 총 합계
traffic_03['총 합계']=traffic_03['승차인원 합계'] + traffic_03['하차인원 합계']
traffic_04['총 합계']=traffic_04['승차인원 합계'] + traffic_04['하차인원 합계']
traffic_05['총 합계']=traffic_05['승차인원 합계'] + traffic_05['하차인원 합계']
```

```python
traffic_03
```

```python
# 총 합계별 교통 혼잡도
count, bin_dividers = np.histogram(traffic_03['총 합계'], bins=3)
bin_names=['여유', '보통', '혼잡']
traffic_03['교통 혼잡도']=pd.cut(x=traffic_03['총 합계'], bins=bin_dividers, labels = bin_names, include_lowest=True)

count, bin_dividers = np.histogram(traffic_04['총 합계'], bins=3)
bin_names=['여유', '보통', '혼잡']
traffic_04['교통 혼잡도']=pd.cut(x=traffic_04['총 합계'], bins=bin_dividers, labels = bin_names, include_lowest=True)

count, bin_dividers = np.histogram(traffic_05['총 합계'], bins=3)
bin_names=['여유', '보통', '혼잡']
traffic_05['교통 혼잡도']=pd.cut(x=traffic_05['총 합계'], bins=bin_dividers, labels = bin_names, include_lowest=True)
```

#### 각 월 별 승하차 Top5, Bottom5

```python
# 3월
subway_03.reset_index(inplace=True)

# 3월 승차 및 하차의 Top5
subway_03_in_h=subway_03.sort_values('승차인원 합계', ascending=False).iloc[:5]
subway_03_out_h=subway_03.sort_values('하차인원 합계', ascending=False).iloc[:5]
subway_03_in_h.rename(columns={'승차인원 합계':'승차인원 Top5'},inplace=True)
subway_03_out_h.rename(columns={'하차인원 합계':'하차인원 Top5'},inplace=True)

# 3월 승차 및 하차의 Bottom5
subway_03_in_l=subway_03.sort_values('승차인원 합계', ascending=False).iloc[-5:]
subway_03_out_l=subway_03.sort_values('하차인원 합계', ascending=False).iloc[-5:]
subway_03_in_l.rename(columns={'승차인원 합계':'승차인원 Bottom5'},inplace=True)
subway_03_out_l.rename(columns={'하차인원 합계':'하차인원 Bottom5'},inplace=True)
```

```python
# 4월
subway_04.reset_index(inplace=True)

# 4월 승차 및 하차의 Top5
subway_04_in_h=subway_04.sort_values('승차인원 합계', ascending=False).iloc[:5]
subway_04_out_h=subway_04.sort_values('하차인원 합계', ascending=False).iloc[:5]
subway_04_in_h.rename(columns={'승차인원 합계':'승차인원 Top5'},inplace=True)
subway_04_out_h.rename(columns={'하차인원 합계':'하차인원 Top5'},inplace=True)

# 4월 승차 및 하차의 Bottom5
subway_04_in_l=subway_04.sort_values('승차인원 합계', ascending=False).iloc[-5:]
subway_04_out_l=subway_04.sort_values('하차인원 합계', ascending=False).iloc[-5:]
subway_04_in_l.rename(columns={'승차인원 합계':'승차인원 Bottom5'},inplace=True)
subway_04_out_l.rename(columns={'하차인원 합계':'하차인원 Bottom5'},inplace=True)
```

#### 각 월 별 승차 하차 구별해서 프레임 만들기

```python
# 3월
subway_03_in_h.drop('하차인원 합계',axis=1,inplace=True)
subway_03_out_h.drop('승차인원 합계',axis=1,inplace=True)
subway_03_in_l.drop('하차인원 합계',axis=1,inplace=True)
subway_03_out_l.drop('승차인원 합계',axis=1,inplace=True)

# 4월
subway_04_in_h.drop('하차인원 합계',axis=1,inplace=True)
subway_04_out_h.drop('승차인원 합계',axis=1,inplace=True)
subway_04_in_l.drop('하차인원 합계',axis=1,inplace=True)
subway_04_out_l.drop('승차인원 합계',axis=1,inplace=True)

# 5월
subway_05_in_h.drop('하차인원 합계',axis=1,inplace=True)
subway_05_out_h.drop('승차인원 합계',axis=1,inplace=True)
subway_05_in_l.drop('하차인원 합계',axis=1,inplace=True)
subway_05_out_l.drop('승차인원 합계',axis=1,inplace=True)
```
