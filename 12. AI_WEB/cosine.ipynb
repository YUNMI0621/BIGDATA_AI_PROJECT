#### 코사인 유사도

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
def make_cosine(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
from konlpy.tag import Kkma
from tqdm import tqdm

KKma = Kkma()

box = []
for i in tqdm(kakao_df['text']):
    if isinstance(i, list):
        i = " ".join(i)  # 리스트를 문자열로 변환
    a = KKma.nouns(i)
    box.append(" ".join(a))

kakao_df['nouns'] = box
```

```python
kakao_df
```

```python
dtm = cv.fit_transform(kakao_df['nouns'])
dtm_df = pd.DataFrame(dtm.toarray(), columns=cv.get_feature_names_out())
dtm_df
```

```python
target = dtm_df.iloc[0].to_numpy()

box = []
for i in range(len(dtm_df)):
    sample = dtm_df.iloc[i].to_numpy()
    cosine = make_cosine(target, sample)
    box.append(cosine)

kakao_df['유사도 분석'] = box
kakao_df.sort_values('유사도 분석', ascending=False, inplace=True) # 내림차순
kakao_df
```

#### 빈도분석

```python
from konlpy.tag import Okt
okt = Okt()
total = []
box = []
text_ = kakao_df['nouns']
box.append(" ".join(text_))
for i in tqdm(okt.nouns(box[0])):
    if len(i)>1:
        total.append(i)

total
```

```python
# 빈도분석
from collections import Counter
result = Counter(total)

dic = {}
for a in tqdm(result.most_common(20)):
    dic[a[0]] = a[1]
dic
```
