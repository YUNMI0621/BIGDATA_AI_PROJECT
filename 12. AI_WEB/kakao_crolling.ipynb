```python
# ë¸Œë¼ìš°ì € ì´ˆê¸°í™”
driver = webdriver.Chrome()
```

```python
gift_name = []
textfull=[]

for h in kakao_df['href']:
    driver.get(h)
    driver.implicitly_wait(5)
    time.sleep(2)

    # # ìŠ¤í¬ë¡¤ ë‚´ë¦¬ê¸°
    # driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)

    # íŒì—…ì°½ ì œê±°

    # ì„ ë¬¼ ì´ë¦„ ê°€ì ¸ì˜¤ê¸°
    gift_name.append(driver.find_element(By.CLASS_NAME, 'tit_subject').text)

    # ì„ ë¬¼í›„ê¸° íƒ­ í´ë¦­
    driver.find_elements(By.CLASS_NAME, 'txt_tab')[1].click()
    time.sleep(2)
    try:
        driver.find_element(By.CSS_SELECTOR, '.inner_layer button').click()
    except:
        print("ERR")
        pass

    # ë”ë³´ê¸° í´ë¦­
    button = driver.find_element(By.CSS_SELECTOR, '.wrap_btn > .btn_more')
    for i in range(0,10):
        for x in range(10):
            try:
                time.sleep(0.5)
                print("ë‚´ë¦¬ê¸° ìš”ì²­")
                button.click()
                break
            except:
                driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)


                pass
        time.sleep(3)

    # ëŒ“ê¸€ í¬ë¡¤ë§
    texts = []
    try:
        for i in range(100):
            texts.append(driver.find_elements(By.CLASS_NAME, 'txt_review')[i].text)
    except:
        pass
    textfull.append(texts)
```
