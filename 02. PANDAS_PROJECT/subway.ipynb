### [1] 데이터 준비

- DF 로딩 / 기본 정보 확인

```python
# DF 로딩
# 데이터 파일 중 2022년 봄에 관한 데이터 추출
subway = pd.read_csv('./서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv', encoding='cp949')
subway_spring = subway[(subway['사용월']>=202203) & (subway['사용월']<=202205)]
subway_spring.reset_index(drop=True, inplace=True)
subway_spring
```

### [2] 데이터 전처리

- 결측치 체크 / 중복 체크

```python
# 결측치 체크
subway_spring.isnull().sum() # 결측치 X
```

```python
# 중복 체크
subway_spring.duplicated().sum() # 중복 X
```

#### 승하차별 합계 구하기

```python
# 승차 합계
subway_in = subway_spring.T.iloc[3:-1:2].sum()
# 하차 합계
subway_out = subway_spring.T.iloc[4:-1:2].sum()
```

```python
subway_spring = subway_spring.T.copy()[:3]
```

```python
subway_spring
```

- 승차인원 합계와 하차인원 합계를 데이터프레임에 합치기

```python
subway_spring.loc['승차인원 합계'] = subway_in
subway_spring.loc['하차인원 합계'] = subway_out
subway_spring = subway_spring.T
```

```python
subway_spring
```

- 각 월별 데이터프레임 만들기

```python
subway_spring = subway_spring.set_index('사용월')

# 3월
subway_03 = subway_spring.loc[202203]
# 4월
subway_04=subway_spring.loc[202204]
# 5월
subway_05=subway_spring.loc[202205]
```

```python
# 확인
subway_03
```

- 총 합계 컬럼 만들기 (승차인원 합계 + 하차인원 합계)

```python
# 3월
subway_03['총 합계'] = subway_03['승차인원 합계'] + subway_03['하차인원 합계']
# 4월
subway_04['총 합계'] = subway_04['승차인원 합계'] + subway_04['하차인원 합계']
# 5월
subway_05['총 합계'] = subway_05['승차인원 합계'] + subway_05['하차인원 합계']
```

```python
# 확인
subway_03
```

#### 총 합계를 기준으로 교통 혼잡도 구간 분할

```python
# 총 합계별 교통 혼잡도
count, bin_dividers = np.histogram(subway_03['총 합계'], bins=3)
bin_names=['여유', '보통', '혼잡']
subway_03['교통 혼잡도']=pd.cut(x=subway_03['총 합계'], bins=bin_dividers, labels = bin_names, include_lowest=True)

count, bin_dividers = np.histogram(subway_04['총 합계'], bins=3)
bin_names=['여유', '보통', '혼잡']
subway_04['교통 혼잡도']=pd.cut(x=subway_04['총 합계'], bins=bin_dividers, labels = bin_names, include_lowest=True)

count, bin_dividers = np.histogram(subway_05['총 합계'], bins=3)
bin_names=['여유', '보통', '혼잡']
subway_05['교통 혼잡도']=pd.cut(x=subway_05['총 합계'], bins=bin_dividers, labels = bin_names, include_lowest=True)
```

```python
# 확인
subway_03
```

```python
subway_04.reset_index(inplace=True)

# 승차 Top5, Bottom5
subway_04_in_top5 = subway_04.sort_values('승차인원 합계', ascending=False).iloc[:5]
subway_04_in_bottom5 = subway_04.sort_values('승차인원 합계', ascending=False).iloc[-5:]

# 하차 Top5, Bottom5
subway_04_out_top5 = subway_04.sort_values('하차인원 합계', ascending=False).iloc[:5]
subway_04_out_bottom5 = subway_04.sort_values('하차인원 합계', ascending=False).iloc[-5:]
```

#### 승하차 인원이 많은 곳의 특징은 일자리, 놀거리로 인한 유동인구가 많은 지역인 것으로 예상된다.

#### 3, 4, 5월 중 승차인원이 많은 역 구하기

```python
subway_03_04_05 = pd.concat([subway_03, subway_04, subway_05])
subway_03_04_05.reset_index(inplace=True)
subway_03_04_05.drop(columns='level_0', inplace=True)
subway_03_04_05.drop(columns='index', inplace=True)
subway_03_04_05
```

```python
subway_03_04_05.sort_values('승차인원 합계', inplace=True, ascending=False)
subway_03_04_05
```

#### 강남과 잠실이 대체적으로 높음. 2022년 4월을 보면 잠실이 강남보다 높은 것을 볼 수 있다. 그 결과에 대한 해석은 코로나 이후 처음 개최된 석촌호수 벚꽃축제로 인한 결과로 보여지고 또한 공연의 빈도도 높아진 것으로 예상된다.

- 월별 승하차 총 합계

```python
# 3월
total_03 = pd.DataFrame([subway_03['승차인원 합계'].sum(), subway_03['하차인원 합계'].sum()])
total_03 = total_03.T

# 4월
total_04 = pd.DataFrame([subway_04['승차인원 합계'].sum(), subway_04['하차인원 합계'].sum()])
total_04 = total_04.T

# 5월
total_05 = pd.DataFrame([subway_05['승차인원 합계'].sum(), subway_05['하차인원 합계'].sum()])
total_05 = total_05.T
```

```python
total_03
```

```python
# 데이터프레임으로 만들기
total = pd.concat([total_03, total_04, total_05])
total.reset_index(drop=True, inplace=True)

# 행, 열 이름 다시 설정
total.rename({0 : '3월', 1 : '4월', 2 : '5월'}, inplace=True)
total.rename(columns={0 : '승차인원 총합계', 1 : '하차인원 총합계'}, inplace=True)

# 승하차 인원 차이 (승차인원 총합계 - 하차인원 총합계)
total['승하차 인원 차이'] = total['승차인원 총합계'] - total['하차인원 총합계']
```

```python
total
```

#### 3월, 4월, 5월 중 승하차인원이 가장 많은 달은 5월이다. 그 이유는 어린이날, 어버이날, 부처님 오신 날 등 각종 행사가 많아서 그런 것으로 생각된다.

#### 승차인원과 하차인원이 같아야 하지만 승하차 인원이 다른 이유는 교통카드를 찍지 않고 나가거나 일회용 교통카드 등을 분실한 경우라고 생각할 수 있다.
